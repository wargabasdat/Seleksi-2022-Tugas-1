```python

import pandas as pd
df = pd.read_csv(r'.\wardah.csv')
```

```python
df
```

1. Visualisasi Data untuk Jumlah Item di Setiap Sub Kategori

```python
df['sub_category'].value_counts().plot(kind = "barh")
```

2. Visualisasi Data untuk Menampilkan 5 Produk Termahal

```python
largest5 = df.nlargest(n=5, columns = ['price'])
largest5.plot.area(x="product_name",y="price",figsize=(18,7))
```

3. Visualisasi Data untuk Menampilkan 5 Produk Termurah

```python
smallest5 = df.nsmallest(n=5, columns = ['price'])
smallest5.plot.line(x="product_name",y="price",figsize=(18,7))
```

```python
df[df['product_name'].str.contains("Foundation")].plot.line(x="product_name", y="price",figsize=(18,10))
```
